```python
#Armstrong number
num = 579
sum = 0
temp = num
l = len(str(num))
#print(l)
for i in range(l+1):
  rem = num%10
  sum = sum+(rem**l)
  num = num//10
print(sum)
if(temp==sum):
  print("Armstrong")
else:
  print("Not a armstrong")
```

**Set**

Definition

A Set is an unordered collection of unique elements.

Features Unordered Mutable No duplicates allowed Faster searching

```python
numbers = {4, 6, 8, 10}
print(numbers)
```

Important Functions Function Description

add() Add element

remove() Remove element

union() Combine sets

intersection() Common elements

difference() Unique elements

clear() Remove all elements

```python
name1 = {'Tharun', 'swamy', 'srinath', 'shiva'}
name2 = {'srinath', 'shiva', 'charan', 'kisan'}
print(name1)
print(name2)
name1.add('Ram')
print(name1)
name1.remove('srinath')
print(name1)
unions = name1.union(name2)
print(unions)
inter = name1.intersection(name2)
print(inter)
dif = name1.difference(name2)
print(dif)
name1.clear()
print(name1)
```

Dictionary

Definition

A Dictionary stores data in key-value pairs.

Features Mutable Keys must be unique Fast lookup Stores related information together

```python
student = {
   "name": "charan",
   "age": 24,
   "course": "AIML"
}
print(student)
print(student.keys())
print(student.values())
print(student.items())
print(student.get('name'))
student.pop('name')
print(student)
student.update({"name": "Nani"})
print(student)
student.update({"age": 25})
print(student)
student.update({"course": "CSE"})
print(student)
```

Important Functions Function Description

keys() Returns all keys

values() Returns all values

items() Returns key-value pairs

get() Safely access value

pop() Remove key

update() Update dictionary

Strings in Python

What is a String?

A String is a sequence of characters enclosed within single quotes (' '), double quotes (" "), or triple quotes (''' ''' or """ """).

Features of Strings

Strings are ordered

Strings are immutable (cannot be modified directly)

Supports indexing and slicing

Can contain letters, numbers, and special characters


String Indexing

Each character has a position called an index.

text = "Python"

Character P y t h o n

Positive Index 0 1 2 3 4 5

Negative Index -6 -5 -4 -3 -2 -1

Function Purpose

upper() Convert to uppercase

lower() Convert to lowercase

capitalize() First letter uppercase

title() Every word first letter uppercase

strip() Remove spaces

replace('replaceval) Replace text

split() Convert string to list

find('str') Find position returns index

count('char) Count occurrences

startswith(value) Check starting text returns bool

endswith('value) Check ending text returns bool
